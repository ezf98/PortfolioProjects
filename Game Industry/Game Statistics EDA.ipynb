```python
#First let's import the libraries
import pandas  as pd
import seaborn as sns
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)
```

```python
#Now let's retrieve and explore the data

df = pd.read_csv(r'C:\Users\ezf98\Downloads\game_statistics_feb_2023.csv\game.csv')
```

```python
df.info()
```

```python
#Let's see the percentage of null values for each column.
for col in df.columns:
    pcntmissing = np.mean(df[col].isnull())
    print('{} - {}'.format(col, pcntmissing))
```

```python
#I will create two csv files, one for sales info and the other for score. It can later be joined in Tableau.
#While reviewing the original csv file, I noticed there are values with 0 instead of null for sales. Let's deal with that first.

df.replace('0.00m', np.nan, inplace=True)
```

```python
df.info()
```

```python
#let's remove the m in the sales too.
```

```python
df['total_sales'] = df['total_sales'].str.rstrip('m')
```

```python
df['japan_sales'] = df['japan_sales'].str.rstrip('m')
df['na_sales'] = df['na_sales'].str.rstrip('m')
df['pal_sales'] = df['pal_sales'].str.rstrip('m')
df['other_sales'] = df['other_sales'].str.rstrip('m')
```

```python
#As we saw above, all this columns have the wrong data type. 
df['total_sales'] = df['total_sales'].astype('float64')
df['japan_sales'] = df['japan_sales'].astype('float64')
df['na_sales'] = df['na_sales'].astype('float64')
df['pal_sales'] = df['pal_sales'].astype('float64')
df['other_sales'] = df['other_sales'].astype('float64')
```

```python
df.info()
```

```python
#Finally for the Sales CSV, we should also remove the null values for this columns.

df1 = df.dropna(subset = ['total_sales','japan_sales','na_sales','pal_sales', 'other_sales'], how= 'all')
```

```python
df1
```

```python
df1 = df1[['title', 'total_sales','japan_sales','na_sales','pal_sales', 'other_sales', 'platform', 'developer']]
```

```python
df1.to_csv(r'C:\Users\ezf98\Game Sales.csv', header = 'column_names')
```
